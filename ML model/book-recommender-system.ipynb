## Collaborative Filtering Based Recommender System

```python
# who reads the books more then 200 they are good_users...........! 
# it is best for our dataset

x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200 
good_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(good_users)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50  # 50 it is best for our dataset

famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

so we have 810 users & 706 books!

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:11]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
# Assuming popular_df has a column 'Book-Title' containing book titles
popular_book_titles = popular_df['Book-Title'].unique()

# Dictionary to store recommendations for each book
all_recommendations = {}

for book_title in popular_book_titles:
    try:
        recommendations = recommend(book_title)
        all_recommendations[book_title] = recommendations
    except Exception as e:
        print(f"Error recommending for book {book_title}: {e}")

# Convert the recommendations dictionary to a DataFrame for easier viewing
recommendations_df = pd.DataFrame([
    {'Book': book, 'Recommendations': recs} for book, recs in all_recommendations.items()
])

recommendations_df
```

```python
# Assuming popular_df has a column 'Book-Title' containing book titles
popular_book_titles = popular_df['Book-Title'].unique()

# List to store all recommendations
all_recommendations = []

for book_title in popular_book_titles:
    try:
        recommendations = recommend(book_title)
        for rec in recommendations:
            all_recommendations.append({
                'Input Book Title': book_title,
                'Recommended Book Title': rec[0],
                'Recommended Book Author': rec[1],
                'Recommended Book Image URL': rec[2]
            })
    except Exception as e:
        print(f"Error recommending for book {book_title}: {e}")

# Convert the list of recommendations to a DataFrame
recommendations_df = pd.DataFrame(all_recommendations)

# Display the resulting DataFrame
recommendations_df
```
